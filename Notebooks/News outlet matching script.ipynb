This script cleans the index/row values to allow for accurate matching of ideological scores across tables

```python
import pandas as pd
import re
import openpyxl
```

```python
# Load the existing Excel file
file_path = '/Users/akilhuang/Documents/SUDS/Updated NELA Master Spreadsheet.xlsx'
workbook = openpyxl.load_workbook(file_path)

# Load the 2018 Orientation worksheet
sheet_name = '2018 Orientation'
worksheet_2018_Orientation = workbook[sheet_name]

sheet = workbook.active
sheet
```

```python
# Convert the 2018 Orientation worksheet data to a DataFrame
data = worksheet_2018_Orientation.values
columns = next(data)
df_2018 = pd.DataFrame(data, columns=columns)
```

```python
# Clean text function
def clean_text(text):
    cleaned_string = []
    for value in text:
        # Convert text to lowercase
        value = str(value).lower()
        # Remove spaces
        value = value.strip()
        value = value.replace(" ", "")
        # Remove newline characters
        value = re.sub(r'\s+', '', value)
        # Remove punctuation
        value = re.sub(r'[^\w\s]', '', value)
        cleaned_string.append(value)
    return cleaned_string
```

```python
# Perform modifications on the selected worksheet
df_2018['source'] = clean_text(df_2018['source'])

# View the dataframe
pd.set_option('display.max_rows', None)
print(df_2018)
```

```python
# Write dataframe back to Excel file
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    df_2018.to_excel(writer, sheet_name=sheet_name, index=False)
```
